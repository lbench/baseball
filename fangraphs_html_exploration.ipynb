```python
import os
import pandas as pd

read_directory = 'fangraphs_htmls'

file_list = sorted(os.listdir(read_directory))

file = file_list[2]

filename = os.fsdecode(file)

to_read = os.path.join(read_directory, filename)

list_of_df = pd.read_html(to_read)
```

```python
print(file_list)
file_list = [name for name in file_list if 'undisclosed' not in name]
print(file_list)
```

```python
for df in list_of_df:
    print(df.columns[0])
```

```python
list_of_df = [ele for ele in list_of_df if "Name" == ele.columns[0]][::2]
print(list_of_df)
```

```python
df = pd.concat(list_of_df)
print(df)
```

```python
lamb = lambda x: x.split()[1] + ', ' + x.split()[0]
df['player_name'] = df['Name'].apply(lamb)
col = list(df.columns)

col.insert(0, col.pop())
df = df[col]
print(df)
```
